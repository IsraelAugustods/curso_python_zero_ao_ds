```python
import pandas as pd 
data = pd.read_csv('datasets/kc_house_data.csv')

pd.set_option('display.float_format',lambda x: '%.2f' % x)
```

```python
#1 Qual o numero de imoveis por ano de construção?

data[['id', 'yr_built']].groupby('yr_built').count()
```

```python
#2 Qual menor numero de quartos por ano de construção de imoveis?

data[['bedrooms', 'yr_built']].groupby('yr_built').min()
```

```python
#3 Qual o preço de compra mais alto por cada numero de quartos?
data[['price', 'bedrooms']].groupby('bedrooms').max()
```

```python
#4 Qual a soma de todos os preços de compra por número de quartos?

data[['price', 'bedrooms']].groupby('bedrooms').sum()
```

```python
import pandas as pd 
from matplotlib import pyplot as plt 
import plotly.express as px 
import plotly.figure_factory as ff
from matplotlib import gridspec
import seaborn as sns
data = pd.read_csv('datasets/kc_house_data.csv')
```

```python
#1 Criar uma nova coluna 'dormitory_type', se utilizada da numero de bedrooms 

data ['dormitory_type'] = data['bedrooms'].apply(lambda x: 'studio' if x == 1 else
                                                           'apartament' if x == 2 else
                                                           'house' if x>2 else 'NA')
```

```python
#2 Faça um grafico de barras que represente a soma dos preços pelo numero de quarto?
```

```python
by_bedrooms = data[['price', 'bedrooms']].groupby ('bedrooms').sum().reset_index()
```

```python
plt.figure(figsize = (20, 10))
plt.bar( by_bedrooms ['bedrooms'] , by_bedrooms ['price'])
```

```python
#3 Faça um grafico de linhas que represente a media dos preços pelo ano de construção dos imoveis. 
```

```python
yr_mean = data[['price', 'yr_built']].groupby('yr_built').mean().reset_index()
```

```python
plt.figure(figsize =(20,12 ))
# plt.plot(yr_mean ['yr_built'], yr_mean ['price'])
sns.lineplot(x= 'yr_built', y= 'price', data = yr_mean)
```

```python
#4 Faça um grafico de barras que represente a media dos preços pelo tipo de dormitorio.
```

```python
mean_dormitory = data[['price', 'dormitory_type']].groupby('dormitory_type').mean().reset_index()
```

```python
plt.figure (figsize = (20,12))
# plt.bar(mean_dormitory ['dormitory_type'], mean_dormitory ['price'])
sns.barplot (x= 'dormitory_type', y= 'price', data = mean_dormitory)
```

```python
#5 Faça um grafico de linhas que mostre a evolução  da media  dos preços pelo ano de reforma dos imoveis, 
#apartir do ano de 1930.
```

```python
df = data.loc[data['yr_renovated'] > 1930,['price', 'yr_renovated']].groupby('yr_renovated').mean().reset_index()
plt.figure (figsize = (20,12))
sns.lineplot (x='yr_renovated', y= 'price', data = df);

```
